#Assignment No-3
Analyzing Sales Data with Matplotlib Suppose you have a CSV file named sales_data.csv containing information about sales data for a company. The dataset includes the following columns:
•	Date: Date of the sale
•	Product: Product name
•	Units_Sold: Number of units sold
•	Revenue: Revenue generated from the sale
Perform the following tasks using Matplotlib:
1. Read the dataset from the CSV file into a DataFrame.
2. Plot a line chart showing the trend of revenue over time (Date).
3. Plot a bar chart showing the total units sold for each product.
4. Plot a pie chart showing the percentage of revenue generated by each product.
5. Customize the plots by adding appropriate labels, titles, and legends.
6. Save each plot as an image file (e.g., PNG format).

You can use the provided dataset sales_data.csv to perform these tasks. Once completed, you will have practiced loading data from a CSV file, creating various types of plots, and customizing plot aesthetics in Matplotlib.


#Solution:

#1. Read the dataset from the CSV file into a DataFrame

```python
# Importing necessary libraries
import matplotlib.pyplot as plt
import pandas as pd

# Read the dataset from the CSV file into a DataFrame
url = '/content/sales_data.csv'
df = pd.read_csv(url)
df.head()
```

#2. Plot a line chart showing the trend of revenue over time (Date)

```python
# Ensure the data types are correct
df['Date'] = pd.to_datetime(df['Date'])  # Convert 'Date' column to datetime type


df_sorted = df.sort_values(by='Date')
plt.figure(figsize=(10, 6))
plt.plot(df_sorted['Date'], df_sorted['Revenue'], color='blue', marker='o', linestyle='-', label='Revenue')
plt.title('Revenue Trend Over Time', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Revenue ($)', fontsize=12)
plt.legend(fontsize=10)
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('revenue_trend_over_time.png')  # Save the plot as an image file
plt.show()
```

#3. Plot a bar chart showing the total units sold for each product

```python
product_units_sold = df.groupby('Product')['Units_Sold'].sum()

plt.figure(figsize=(10, 6))
product_units_sold.plot(kind='bar', color='orange', alpha=0.8, edgecolor='black')
plt.title('Total Units Sold for Each Product', fontsize=16)
plt.xlabel('Product', fontsize=12)
plt.ylabel('Units Sold', fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('total_units_sold.png')  # Save the plot as an image file
plt.show()
```

#4. Plot a pie chart showing the percentage of revenue generated by each product

```python
product_revenue = df.groupby('Product')['Revenue'].sum()

plt.figure(figsize=(8, 8))
product_revenue.plot(kind='pie', autopct='%1.1f%%', startangle=90,
                     colors=plt.cm.Paired.colors, textprops={'fontsize': 10})
plt.title('Percentage of Revenue by Product', fontsize=16)
plt.ylabel('')  # Hide the y-axis label
plt.tight_layout()
plt.savefig('revenue_percentage_by_product.png')  # Save the plot as an image file
plt.show()
```

#5. Customize the plots by adding appropriate labels, titles, and legends.

As we can see, all the plots above are customized by adding appropriate labels, titles, and legends.

#6. Save each plot as an image file (e.g., PNG format).
